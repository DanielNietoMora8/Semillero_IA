```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import urllib
import numpy as np


filename = "/content/drive/MyDrive/xeno-canto/XC761948 - Tawny Owl - Strix aluco.mp3"
y, sr = librosa.load(filename)
# Extraer una porción de 5 segundos de la grabación de canto de ave
start_time = 10.0 # En segundos
end_time = 15.0 # En segundos
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)
bird_call = y[start_sample:end_sample]

# Mostrar un espectograma de la porción extraída
"""
Esta sección utiliza la función librosa.feature.melspectrogram para calcular 
un espectograma de la porción de canto de ave extraída. Luego, utiliza la función 
librosa.display.specshow para mostrar el espectograma con matplotlib. La función 
librosa.power_to_db convierte la energía del espectograma en dB y el parámetro ref=np.max 
establece la referencia máxima en el máximo valor del espectograma. La función plt.colorbar agrega una 
barra de colores que muestra la escala de dB y plt.title agrega un título

"""
plt.figure(figsize=(12, 4))
librosa.display.specshow(librosa.power_to_db(librosa.feature.melspectrogram(bird_call, sr=sr), ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectograma del canto de ave')
plt.tight_layout()
plt.show()
```

```python


# Estimar el tempo de la grabación
tempo = librosa.beat.tempo(y, sr=sr)

# Calcular los coeficientes cepstrales de mel-frecuencia (MFCC)
mfccs = librosa.feature.mfcc(y, sr=sr)

# Calcular la croma de la grabación
chroma = librosa.feature.chroma_stft(y, sr=sr)

# Calcular el espectograma de mel-frecuencia de la grabación
melspec = librosa.feature.melspectrogram(y, sr=sr)

# Visualizar la forma de onda de la grabación
plt.figure(figsize=(12, 4))
librosa.display.waveplot(y, sr=sr)
plt.title('Forma de onda del audio')
"""

El espectograma de mel-frecuencia es una 
representación gráfica de la frecuencia y la amplitud de una señal 
de audio en términos de la escala de mel. La escala de mel se basa en la 
percepción humana de la frecuencia y es logarítmica en lugar de lineal, lo que significa 
que se corresponde más con la forma en que el oído humano percibe la frecuencia.
"""

# Visualizar el espectograma de mel-frecuencia
plt.figure(figsize=(12, 4))
librosa.display.specshow(librosa.power_to_db(melspec, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectograma de mel-frecuencia')

plt.show()


print("Tempo: {:.2f} beats por minuto".format(tempo[0]))
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import urllib
import numpy as np


filename = "/content/drive/MyDrive/xeno-canto/XC761948 - Tawny Owl - Strix aluco.mp3"
y, sr = librosa.load(filename)
# Extraer una porcin de 5 segundos de la grabacin de canto de ave
start_time = 10.0 # En segundos
end_time = 15.0 # En segundos
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)
bird_call = y[start_sample:end_sample]

# Mostrar un espectograma de la porcin extrada
"""
Esta seccin utiliza la funcin librosa.feature.melspectrogram para calcular 
un espectograma de la porcin de canto de ave extrada. Luego, utiliza la funcin 
librosa.display.specshow para mostrar el espectograma con matplotlib. La funcin 
librosa.power_to_db convierte la energa del espectograma en dB y el parmetro ref=np.max 
establece la referencia mxima en el mximo valor del espectograma. La funcin plt.colorbar agrega una 
barra de colores que muestra la escala de dB y plt.title agrega un ttulo

"""
plt.figure(figsize=(12, 4))
librosa.display.specshow(librosa.power_to_db(librosa.feature.melspectrogram(bird_call, sr=sr), ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectograma del canto de ave')
plt.tight_layout()
plt.show()
```

```python


# Estimar el tempo de la grabacin
tempo = librosa.beat.tempo(y, sr=sr)

# Calcular los coeficientes cepstrales de mel-frecuencia (MFCC)
mfccs = librosa.feature.mfcc(y, sr=sr)

# Calcular la croma de la grabacin
chroma = librosa.feature.chroma_stft(y, sr=sr)

# Calcular el espectograma de mel-frecuencia de la grabacin
melspec = librosa.feature.melspectrogram(y, sr=sr)

# Visualizar la forma de onda de la grabacin
plt.figure(figsize=(12, 4))
librosa.display.waveplot(y, sr=sr)
plt.title('Forma de onda del audio')
"""

El espectograma de mel-frecuencia es una 
representacin grfica de la frecuencia y la amplitud de una seal 
de audio en trminos de la escala de mel. La escala de mel se basa en la 
percepcin humana de la frecuencia y es logartmica en lugar de lineal, lo que significa 
que se corresponde ms con la forma en que el odo humano percibe la frecuencia.
"""

# Visualizar el espectograma de mel-frecuencia
plt.figure(figsize=(12, 4))
librosa.display.specshow(librosa.power_to_db(melspec, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectograma de mel-frecuencia')

plt.show()


print("Tempo: {:.2f} beats por minuto".format(tempo[0]))
```

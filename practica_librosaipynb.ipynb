```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import urllib
import numpy as np


# Descargar una grabacin de canto de ave desde Xeno-Canto
#url = "https://www.xeno-canto.org/api/2/recordings?query=bearded%20bellbird%20q:A"
filename = "/content/drive/MyDrive/xeno-canto/XC761948 - Tawny Owl - Strix aluco.mp3"
#urllib.request.urlretrieve(filename)

# Cargar la grabacin en memoria con librosa
y, sr = librosa.load(filename)

# Extraer una porcin de 5 segundos de la grabacin de canto de ave
start_time = 10.0 # En segundos
end_time = 15.0 # En segundos
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)
bird_call = y[start_sample:end_sample]

# Mostrar un espectograma de la porcin extrada
"""
Esta seccin utiliza la funcin librosa.feature.melspectrogram para calcular 
un espectograma de la porcin de canto de ave extrada. Luego, utiliza la funcin 
librosa.display.specshow para mostrar el espectograma con matplotlib. La funcin 
librosa.power_to_db convierte la energa del espectograma en dB y el parmetro ref=np.max 
establece la referencia mxima en el mximo valor del espectograma. La funcin plt.colorbar agrega una 
barra de colores que muestra la escala de dB y plt.title agrega un ttulo

"""
plt.figure(figsize=(12, 4))
librosa.display.specshow(librosa.power_to_db(librosa.feature.melspectrogram(bird_call, sr=sr), ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Espectograma del canto de ave')
plt.tight_layout()
plt.show()
```

```python
# Anlisis de tono
#tuning = librosa.tuning.estimate_tuning(bird_call, sr=sr)
#print("La frecuencia fundamental estimada es de {:.2f} Hz".format(tuning))

# Anlisis de ritmo
tempo = librosa.beat.tempo(bird_call, sr=sr)
print("El tempo estimado es de {:.2f} BPM".format(tempo))

# Extraccin de caractersticas
mfccs = librosa.feature.mfcc(bird_call, sr=sr)
chroma = librosa.feature.chroma_stft(bird_call, sr=sr)
spectrogram = librosa.feature.melspectrogram(bird_call, sr=sr)

# Visualizacin
plt.figure(figsize=(12, 4))
plt.subplot(2, 2, 1)
librosa.display.waveplot(bird_call, sr=sr)
plt.title("Seal de audio")

plt.subplot(2, 2, 2)
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
plt.title("MFCCs")

plt.subplot(2, 2, 3)
librosa.display.specshow(chroma, sr=sr, x_axis='time')
plt.title("Chroma")

plt.subplot(2, 2, 4)
librosa.display.specshow(spectrogram, sr=sr, x_axis='time')
plt.title("Espectograma de Mel")

plt.tight_layout()
plt.show()
```
